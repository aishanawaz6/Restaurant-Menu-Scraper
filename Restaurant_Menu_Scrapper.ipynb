```python
# AISHA NAWAZ 
# Task 2: Industry project
# Looking for someone to write a script I can use over and over to scrape a restaurant menu
# and all the modifiers and prices. Prepare a csv for the data.
# https://demoazulmexicanosummerville.onlineordersnow.com/MenuDisplay

#WEB SCRAPPING WITH DYNAMIC CONTENT
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import requests
from bs4 import BeautifulSoup

url= 'https://demoazulmexicanosummerville.onlineordersnow.com/MenuDisplay'
driver = webdriver.Chrome() #Make Sure Driver is in the same directory
driver.get(url)

# Waiting for the View Menu button to be visible
viewMenuButton = WebDriverWait(driver, 30).until(EC.visibility_of_element_located((By.XPATH, "//button[contains(., 'View Menu')]")))

# Clicking the View Menu button
viewMenuButton.click()

# Waiting for the open Menu button to be visible
openMenuButton = WebDriverWait(driver, 30).until(EC.visibility_of_element_located((By.XPATH, "//button[contains(., 'Open Menu')]")))

# Clicking the Open Menu button
openMenuButton.click()

driver.implicitly_wait(30)  # Waiting for the whole dynamic content to load
html = driver.page_source  # Saving the page source [one without the last button clicked]

# Waiting for the size/price buttons to be visible
button = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.CLASS_NAME, "sizeStyle")))

# Clicking the button : (Click for Size/Price) using javaScript
driver.execute_script("arguments[0].click();", button)

driver.implicitly_wait(30)  # Waiting for the whole dynamic content to load
html2 = driver.page_source  # Saving the page source  [one with all buttons clicked]


driver.quit()              # Closing the web driver
```

```python
# Parse HTML 
soup = BeautifulSoup(html, 'html.parser')
soup2 = BeautifulSoup(html2, 'html.parser')

#NOTE: TWO DIFFERENT HTMLS WERE SAVED TO AVOID REDUNDANT AND UNNECCARY DATA SCRAPPING
```

```python
# Finding and extracting Item Names (THAT HAVE DIRECT PRICES)
items = soup.findAll('span', class_='itemLabel')
itemNames=[]
for name in items:
    if name:
        data = name.text.strip()
        if(data!='8" Sm.' and data!='10" Med.' and data!='14" Lg.'): #These were observed to not be itemNames hence removed
            itemNames.append(data)
```

```python
itemNames
```

```python
# Parse HTML 
soup = BeautifulSoup(html, 'html.parser')

# Finding and extracting Item Names (THAT DO NOT HAVE DIRECT PRICES)
items2 = soup.findAll('span', class_='menuItemGroupLabel')
itemNames2=[]
for name2 in items2:
    if name2:
        data1 = name2.text.strip()
        itemNames2.append(data1)
```

```python
itemNames2
```

```python
# Finding and extracting item prices [FOR ITEMS WITH DIRECT PRICES]
itemPRICES = soup2.findAll('span', class_='priceStyle moneyStyle mr-1 floatRight')
prices=[]
for price in itemPRICES:
    if price:
        data2 = price.text.strip()
        prices.append(data2)
```

```python
prices
```

```python
# Finding and extracting item prices [FOR ITEMS WITHOUT DIRECT PRICES]
itemPRICES2 = soup.findAll('div', class_='priceStyle moneyStyle mt-3')
prices2=[]
for price2 in itemPRICES2:
    if price2:
        p = price2.text.strip()
        prices2.append(p)
```

```python
prices2
```

```python
# Finding and extracting all items descriptions [THAT HAVE DIRECT PRICES]
descriptionHTML = soup.findAll('div', class_='itemDescription pt-2 pl-3')
descriptions=[]
for description in descriptionHTML:
    if description:
        data3 = description.text.strip()
        descriptions.append(data3)
```

```python
descriptions
```

```python
# Finding and extracting all items descriptions [OF ITEMS WITHOUT DIRECT PRICES]
descriptionHTML2 = soup.findAll('div', class_='itemDescription pt-2 pb-4 pl-2 clickable')
descriptions2=[]
for description2 in descriptionHTML2:
    if description2:
        data4 = description2.text.strip()
        descriptions2.append(data4)
```

```python
descriptions2
```

```python
print('COUNTING ITEMS--> ITEMS 1 (with direct prices) & ITEMS 2 (without direct prices)')
print('ITEMS 1: ',len(itemNames),'PRICES 1:',len(prices),'DES1:',len(descriptions),'ITEMS 2:',len(itemNames2),'DES2:',len(descriptions2))
```

```python
rows=[['Item Name','Description','Price']]

for i in range (len(itemNames)): #Adding to list items with direct prices
    rows.append([itemNames[i],descriptions[i],prices[i]])

for j in range (len(itemNames2)): #Adding to list items without direct prices
    rows.append([itemNames2[j],descriptions2[j],'PRICES VARY ACCORDING TO SIZE: '+str(prices2)])
```

```python
rows
```

```python
#SAVING SCRAPPED DATA TO A CSV FILE:

import openpyxl
workbook = openpyxl.Workbook() #A new workbook created
sheet = workbook.active        #Selecting the active sheet

for i, row in enumerate(rows, start=1): # Saving the list items to the workbook
    for j, value in enumerate(row, start=1):
        sheet.cell(row=i, column=j).value = value

workbook.save('RestaurantMenu.xlsx') #Saving the whole workbook to an csv file
```
